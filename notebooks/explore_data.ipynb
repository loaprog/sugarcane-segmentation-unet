```python
import os
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
import argparse

caminho_arquivo = "dados/blocos/bloco_0_1408.png"
imagem_rgb = Image.open(caminho_arquivo)
imagem_rgb = imagem_rgb.convert('RGB')

array_rgb = np.array(imagem_rgb)
R = array_rgb[:, :, 0].astype(float)
G = array_rgb[:, :, 1].astype(float)
B = array_rgb[:, :, 2].astype(float)

gli = (2 * G + R + B) / (2 * G - R - B)

fg, ax = plt.subplots(1, 2, figsize=(12, 6))
ax[0].imshow(imagem_rgb)
ax[0].set_title('RGB')
ax[0].axis('off')
cax = ax[1].imshow(gli, cmap='RdYlGn', vmin=0, vmax=1)
ax[1].set_title('GLI')
ax[1].axis('off')

plt.show()
```
